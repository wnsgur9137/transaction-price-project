```python
year_gu = train_seoul.groupby(["gu","year"])['log_price'].mean()
year_gu = year_gu.reset_index()
year_gu = year_gu.set_index("year")
```

```python
gu_name = year_gu.reset_index()['gu'].unique()
```

```python
def gu_linePlot(year_gu, gu_name):
    gu_info_name = {}
    for name in gu_name:
        gu_info = year_gu.loc[year_gu['gu'] == name]
        gu_info = gu_info.drop("gu", axis=1)
        gu_info_name[name] = gu_info
    return gu_info_name

gu_list = gu_linePlot(year_gu, gu_name)
```

```python
import seaborn as sns
for name, data in gu_list.items():
    ax = sns.lineplot(data= data, x='year', y='log_price', label=name)

ax.figure.set_size_inches(15,15)
ax.legend(loc="upper left")
ax.figure.show()
```

## 정규화

```python
import numpy as np
x =  train_seoul['log_price'].values
normalization = (x - x.mean()) / x.std()
normalization = np.sort(normalization)
```

```python
y = (1 / np.sqrt(2 * np.pi)) * np.exp(- normalization ** 2 / 2 )
y
```

```python
plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
plt.plot(normalization, y)
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution without scipy")     # 타이틀 표시
plt.legend(["N(0, 1)"])              # 범례 표시
plt.show()
```
